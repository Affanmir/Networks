```python

import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from matplotlib.lines import Line2D


order = [	"Bantwa","Dhoraji",	"Gondal",	"Jaipur",	"Jetpur",	"Kutchi",	"Morbi Tankara"	,"Mumbai",	"Non-memon",	"Upleta",	"Wasawad",	"Verawal"]


m_stat = 'What is your current marital status?'
partner_jamat = 'Does your partner belong to the Memon community? If Memon, please specify subgroup/jamat.1'
my_jamat = 'What sub-group(jamaat) of the Memon community do you belong to ?'
```

```python
married_df = df.loc[df[m_stat] == 'Married']
married_df[my_jamat] = married_df[my_jamat].str.replace('Dhoraji waley','Dhoraji')
married_df[my_jamat] = married_df[my_jamat].str.replace('Jaipur waley','Jaipur')
married_df[my_jamat] = married_df[my_jamat].str.replace('Opleta','Upleta')
married_df[my_jamat] = married_df[my_jamat].str.replace('Wasawad ','Wasawad')
married_df[my_jamat] = married_df[my_jamat].str.replace('Gondal waley','Gondal')
married_df[my_jamat] = married_df[my_jamat].str.replace('Kutchi memon','Kutchi')
married_df[my_jamat] = married_df[my_jamat].str.replace('Mumbai- waley','Mumbai')
married_df[my_jamat] = married_df[my_jamat].str.replace('Verawal waley','Verawal')



married_df[partner_jamat] = married_df[partner_jamat].str.replace('Dhoraji waley','Dhoraji')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Jaipur waley','Jaipur')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Mumbai waley','Mumbai')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Opleta','Upleta')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Upleta ','Upleta')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('veraval','Verawal')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('verawal','Verawal')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Verawal waley','Verawal')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Bombay waley','Mumbai')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Wasawad ','Wasawad')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('Morbi Tankara Memon Jamaat','Morbi Tankara')
married_df[partner_jamat] = married_df[partner_jamat].str.replace('No','Non-Memon')

My_jammat = married_df[my_jamat]
Partner_jamat = married_df[partner_jamat]


print(married_df[my_jamat].unique())
print(married_df[partner_jamat].unique())


table =  pd.crosstab(Partner_jamat, My_jammat, dropna=False)
table = table.reindex(table.index, axis=1, fill_value=0)
table
```

```python
new_df = table
for i in table.columns:
  for j in table.columns:
    if i==j:
      new_df[i][j] = 1
    new_df[i][j] = new_df[i][j] + new_df[j][i]

for i in table.columns:
  for j in table.columns:
    if new_df[i][j] >= 1:
      new_df[i][j] = 1
  
  
 
 
rows, cols = np.where(new_df >= 1)
edges = zip(rows.tolist(), cols.tolist())
gr = nx.Graph()
gr.add_edges_from(edges)
labelz= {}
for i in range(len(table.columns)):
    labelz[i] = table.columns[i] 
# pos=nx.spring_layout(gr) #default to scale=1
# nx.draw(gr,pos)
# pos=nx.spring_layout(gr,scale=3) # double distance between all nodes
# nx.draw(gr,pos)
 
#nx.draw_networkx_labels(gr, pos, labels, font_size=16)
pos = nx.spring_layout(gr, k=0.25, iterations=15)
nx.draw(gr,pos,arrowsize = 20, node_size=500,labels= labelz,  with_labels=True)
 
centrality = nx.eigenvector_centrality(gr)
centralities ={}
for i in centrality:
  centralities[order[i]] = centrality[i]
print(centralities)
 
bet_cen=nx.betweenness_centrality(gr, k=None, normalized=True, weight=None, endpoints=False, seed=None)
bet_cens = {}
for i in bet_cen:
  bet_cens[order[i]] = bet_cen[i]
print(bet_cens)
 
clo_cen= nx.closeness_centrality(gr, u=None, distance=None, wf_improved=True)
clo_cens = {}
for i in clo_cen:
  clo_cens[order[i]] = clo_cen[i]
print(clo_cens)
 
deg_cen=nx.degree_centrality(gr)
deg_cens = {}
for i in deg_cen:
  deg_cens[order[i]] = deg_cen[i]
print(deg_cens)
#Married People Graph
```
